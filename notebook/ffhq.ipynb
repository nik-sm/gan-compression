```python
from PIL import Image
import random
import os
import matplotlib.pyplot as plt
from torchvision import transforms
```

```python
root_path = '/home/john/projects/ffhq-dataset/images1024x1024'
num_list = random.sample(range(0, 70000), 6)
print(num_list)
```

```python
transform = transforms.Compose([
    transforms.Lambda(lambda img: transforms.functional.crop(img,50,50,900,900)),
    transforms.Resize((128, 128))])
# Can't crop too much on FFHQ

fig, axs = plt.subplots(2, len(num_list), figsize=(30, 10))
for i in range(len(num_list)):
    p = os.path.join(root_path, str(num_list[i]//1000 * 1000).zfill(5), str(num_list[i]).zfill(5) + ".png")
    img = Image.open(p).convert('RGB')
    axs[0, i].imshow(img)

    axs[1, i].imshow(transform(img))
```
